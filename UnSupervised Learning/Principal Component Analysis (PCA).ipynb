```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
data = pd.read_csv("D:/datasets/data_cancer.csv")
```

```python
data
```

```python
X=data.drop(["Unnamed: 32","id","diagnosis"],axis=1)
```

```python
y=data["diagnosis"]
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_transformed=sc.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
pc = PCA(n_components=2)
```

```python
pc.fit(X_transformed)
```

```python
X_Components= pc.fit_transform(X_transformed)
```

```python
X_Components[0]
```

```python
pc.explained_variance_ratio_
```

```python
#Elbow Method
```

```python
len(X.columns)
```

```python
k_value = range(1,len(X.columns)+1)

explained_variance = []

for k in k_value:
    pca =PCA(n_components=k)
    pca.fit(X_transformed)
    explained_variance.append(sum(pca.explained_variance_ratio_))
```

```python
explained_variance
```

```python
plt.plot(k_value,explained_variance)
plt.xlabel("No. of Components")
plt.ylabel("Explained Variance")
plt.title("Elbow Method : To determine value of K")
plt.show()
```
